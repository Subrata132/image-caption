```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
text_file = open('../data/captions.txt', 'r')
text = text_file.read()
text_file.close()
```

```python
lines = text.split('\n')
img_cap_list = []
for line in lines[1:]:
    image_name, caption_list = line.split(',')[0].strip(), line.split(',')[1:]
    caption = ''.join(caption_list).strip()
    img_cap_list.append((image_name, caption))
img_cap_df = pd.DataFrame(img_cap_list, columns=['image_name', 'caption'])
img_cap_df.reset_index()
img_cap_df.head(10)
```

```python
image_names = img_cap_df['image_name'].unique()
```

```python
image_name = image_names[32]
captions = img_cap_df[img_cap_df['image_name']==image_name]['caption'].to_list()
```

```python
fig = plt.figure(figsize=(10, 20))
for i in range(5):
    image_name = image_names[i]
    captions = img_cap_df[img_cap_df['image_name']==image_name]['caption'].to_list()
    ax = fig.add_subplot(5, 2, i*2+1, xticks=[], yticks=[])
    image = plt.imread('../data/Images/'+image_name)
    ax.imshow(image)
    ax = fig.add_subplot(5, 2, i*2+2)
    plt.axis('off')
    ax.plot()
    ax.set_xlim(0,1)
    ax.set_ylim(0,len(captions))
    for i, caption in enumerate(captions):
        ax.text(0,i,caption,fontsize=20)
plt.show()
```
