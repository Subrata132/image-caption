```python
import json
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
with open('plot_data.json') as json_file:
    data = json.load(json_file)
```

```python
data.keys()
```

```python
loss = data['loss']
bleu = data['val_bleu']
```

```python
epochs = np.arange(0, len(loss), 1)
```

```python
plt.figure(figsize=(10, 6))
plt.tight_layout()
plt.plot(epochs, loss, '-o', linewidth=2, color='black', label='Training loss')
plt.grid()
plt.title('Training loss vs. epoch', fontweight='bold', fontsize=16)
plt.xlabel('Epoch',  fontweight='bold', fontsize=16)
plt.ylabel('Loss',  fontweight='bold', fontsize=16)
plt.legend(loc='best')
plt.savefig('loss.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.tight_layout()
plt.plot(epochs, bleu, '-o', linewidth=2, color='black', label='BLEU score for validation set')
plt.title('BLEU score vs. epoch', fontweight='bold', fontsize=16)
plt.xlabel('Epoch',  fontweight='bold', fontsize=16)
plt.ylabel('BLEU score',  fontweight='bold', fontsize=16)
plt.grid()
plt.legend(loc='best')
plt.savefig('bleu_valid.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
with open('test_blue_data.json') as json_file:
    test_data = json.load(json_file)
```

```python
test_data.keys()
```

```python
test = test_data['test_set']
```

```python
red_square = dict(markerfacecolor='r', marker='o')
boxprops = dict(linestyle='-', linewidth=4, color='black')
plt.figure(figsize=(7, 10))
plt.tight_layout()
plt.boxplot(test, flierprops=red_square, boxprops=boxprops)
plt.title('BLEU score', fontweight='bold', fontsize=16)
plt.grid()
plt.xticks([])
plt.savefig('bleu_test.png', dpi=150, bbox_inches='tight')
plt.show()
```
